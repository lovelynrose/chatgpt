# List all models currently accessible

```python
import os
import openai
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
models = openai.Model.list()
```

```python
print(models)
```

```python
print(models["data"][0]["id"])
```

```python
for model in models["data"]:
    print(f"Model ID: {model.id}")
    print(f"Model Name: {model.permission}")
```

# List models that allow fine tuning

```python
# Filter the models that allow fine-tuning
fine_tuning_models = [
    model for model in models['data'] 
    if model.permission[0]['allow_fine_tuning']
]

# Print the fine-tuning models
for model in fine_tuning_models:
    print(model['id'])
```
